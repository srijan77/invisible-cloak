```python
import numpy as np
import cv2
import time
cap=cv2.VideoCapture(0)
#setting the initial variable
background=0
#giving the camera some time to sleep
time.sleep(5)
#Reading the background
while True:
    cv2.waitKey(1000)
    ret,background=cap.read()
    if ret:
        break
background=np.flip(background,1)
#taking the actual frame
while (cap.isOpened()):
    ret,image=cap.read()
    if not ret:
        break
    image=np.flip(image,1)
    
#Creating the mask and detecting the color after converting to hsv
#mask1
    hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
    upper_red_hsv=np.array([100,255,255]) 
    lower_red_hsv=np.array([100,40,40])
    mask1=cv2.inRange(hsv,lower_red_hsv,upper_red_hsv)
#mask2    
    lower_red = np.array([155, 40, 40]) 
    upper_red = np.array([180, 255, 255]) 
    mask2 = cv2.inRange(hsv, lower_red, upper_red) 
    
    mask3=mask1+mask2
 #creating a kernel to remove noise from white background   
    kernel=np.ones((3,3),dtype=np.uint8)
    opening=cv2.morphologyEx(mask3,cv2.MORPH_OPEN,kernel,iterations=2)
    dil_mask = cv2.dilate(opening, kernel, iterations = 1) 
    new_mask = cv2.bitwise_not(dil_mask)
    
    res1=cv2.bitwise_and(background,background,mask=dil_mask)
    res2=cv2.bitwise_and(image,image,mask=new_mask)
    
    res=cv2.addWeighted(res1,0.5,res2,0.5,0)
    
    cv2.imshow("INVISIBLE",res)
    
    if cv2.waitKey(20) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()











#For the blue cloth the value will be:
 #low_blue_hsv_for_mask_1= np.array([94, 80, 30])
 #high_blue_hsv_for_mask_1 = np.array([126, 255, 255])

```
